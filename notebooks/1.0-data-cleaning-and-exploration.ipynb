# Data Cleaning and Initial Exploration

In this notebook we'll do an initial exploration of the Spotify Dataset, which containg ~30.000 songs. We'll check null values, inconsistensies and prepare the data for the next step (Exploratory Data Analysis).

```python
import pandas as pd
```

## Importing the data

```python
file_path = '../backend/data/spotify_songs_dataset.csv'

df = pd.read_csv(file_path)

df.head()
```

## Null data inspection

```python
null_rows = df[df['track_name'].isnull() | df['track_artist'].isnull() | df['track_album_name'].isnull()]

print(f'Rows with null data: {len(null_rows)}')
display(null_rows)
```

The null rows are less than the 0.01% of the total data, so we'll simply drop those rows.

```python
df = df.dropna()

print(f'Dimension without nulls: {df.shape}')
```

## Duplicates

How many duplicated tracks are present in the dataset? We can use `track_id` to check this.

```python
num_duplicates = df.duplicated(subset=['track_id']).sum()
print(f'Number of duplicated tracks: {num_duplicates}')
```

```python
duplicates_sample = df[df.duplicated(subset=['track_id'], keep=False)].sort_values(by='track_id')

print('Sample of duplicated tracks (Same track but different playlist):\n')
print((duplicates_sample[['track_name', 'track_artist', 'playlist_name', 'track_id']]).head(7))
```

Before drop the duplicates, we can create a new popularity metric (in how many playlist this song appears?) and store it in the dataset.

```python
playlist_counts = df['track_id'].value_counts().reset_index()

playlist_counts.columns = ['track_id', 'playlist_appearance_count']

df = df.merge(playlist_counts, on='track_id', how='left')

# playlist_counts.head()
df.head()
```

Now we can drop the duplicates without losing this 'popularity' information of each track.

## Drop the duplicates

```python
df_cleaned = df.drop_duplicates(subset=['track_id'], keep='first').copy()
print(f'Dimension without nulls and duplicates: {df_cleaned.shape}')
```

# Exporting the cleaned dataset

```python
df_cleaned.to_csv('../backend/data/cleaned_dataset.csv', index=False)
```
