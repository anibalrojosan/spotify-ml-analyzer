# 2.0 Exploratory Data Analysis

**Objective:** The primary goal of this notebook is to transition from raw audio features to interpretable psychological insights. We aim to understand how physical sound properties (frequency, amplitude, rhythm) translate into human emotional states and personality traits. This analysis will serve as the foundation for the **Psychological Profile Engine** in the final web application.


Dataset Overview:

* **Volume:** ~30,000 unique tracks.
* **Key Features:** Audio attributes provided by Spotify's API (Valence, Energy, Danceability, Acousticness, etc.).


Analysis Roadmap:

We will focus on four key dimensions to build a comprehensive user profile:

* **A. Statistical Sanity Check (Correlations):**
    * Validating relationships between variables (e.g., *Energy* vs. *Loudness*) to ensure data consistency.
    * identifying redundant features to reduce dimensionality.

* **B. Affective Mapping (The Russell's Circumplex Model):**
    * Plotting **Energy (Arousal)** vs. **Valence (Positivity)**.
    * **Goal:** To segment tracks into emotional quadrants (e.g., High Energy + Low Valence = *Anxious/Aggressive* vs. High Energy + High Valence = *Joyful/Excited*).

* **C. Cognitive vs. Physical Engagement:**
    * Analyzing **Instrumentalness** vs. **Speechiness** to determine if the user prefers narrative-driven content (Lyrics) or abstract processing (Music).
    * Analyzing **Danceability** and **Tempo** to gauge the user's preference for physical regulation and movement.

* **D. Intensity & texture:**
    * Examining **Acousticness** and **Loudness** distributions to understand preferences for organic vs. synthetic sounds. 


**Expected Outcome:** By the end of this notebook, we will identify the specific clusters and thresholds needed to engineer the **`mood_category`** and **`personality_type`** features for the backend API.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
file_path = '../backend/data/cleaned_dataset.csv'
df = pd.read_csv(file_path)

df.info()
```

## Statistical sanity check

Does the data makes sense? It's useful to make predictions and try to interpret user's emotions? Let's figure it out.

1. Audio features: 
    - `Loudness` it's measured in db and should be in the range of [-60, 0]. Positive small numbers can exist if the track it's highly compressed.
    - `Key` it's a categorical feature in Pitch Class notation, it's range should be [0, 11].  
    - All the other features except `tempo` (rpm) and `duration_ms` should be in a range of [0, 1].
    - For this analysis, we'll consider a `duration_ms` of at least be 30000 (30 seconds) because even grincore/punk track commonly have a lenght longer than ~30 seconds. 

```python
audio_features = [
    'danceability', 'energy', 'key', 'loudness', 'mode', 
    'speechiness', 'acousticness', 'instrumentalness', 
    'liveness', 'valence', 'tempo', 'duration_ms'
]

df_audio = df[audio_features]

description = df_audio.describe().T
print(description[['min', 'max', 'mean', 'std']])
```

```python
# Positive loudness
loud_tracks = df[df['loudness'] >= 0].copy()
print(f'Songs with loudness > 0 db: {len(loud_tracks)}')

# Noise cleaning
min_duration_ms = 30_000
df_clean_duration = df[df['duration_ms'] <= min_duration_ms].copy()
print(f'Number of short tracks: {len(df_clean_duration)}')

# Update the cleaned dataframe
df = df[df['duration_ms'] > min_duration_ms].copy()

print(f'Cleaned dataset size: {len(df)}')
```

## Correlation matrix

```python
plt.figure(figsize=(12, 10))

correlation_matrix = df[audio_features].corr()

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Heatmap
sns.heatmap(correlation_matrix, 
            annot=True,         
            fmt=".2f",          
            cmap='coolwarm',    
            mask=mask,          
            vmax=1, vmin=-1,    
            center=0,
            square=True, 
            linewidths=.5)

plt.title('Correlation matrix of audio features')
plt.show()
```

Only Loudness and Energy have a strong positive relationship, and Acousticness and Energy have a strong negative relationships. Neither it's strong enough to delete one due to redundancy. It's interesting to note the week correlation between Valence and Energy, and Valence and Danceability

## Affective Mapping

```python
# We map 'playlist_genre' to color (hue) to observe if genres form distinct emotional clusters
sns.scatterplot(
    data=df, 
    x='valence', 
    y='energy', 
    hue='playlist_genre', 
    alpha=0.6, 
    palette='bright',
    s=30 
)

# Quadrant Thresholds (Based on Russell's Circumplex Model)
# We use 0.5 as the standard neutral baseline for normalized features
plt.axvline(0.5, color='black', linestyle='--', linewidth=1.5)
plt.axhline(0.5, color='black', linestyle='--', linewidth=1.5)

# Quadrant Annotations (Psychological Interpretation)
# Q1: High Arousal, Negative Valence
plt.text(0.05, 0.9, 'AGGRESSIVE / ANXIOUS\n(High Energy, Low Valence)', 
         fontsize=10, backgroundcolor='white', fontweight='bold')

# Q2: High Arousal, Positive Valence
plt.text(0.75, 0.9, 'JOYFUL / EUPHORIC\n(High Energy, High Valence)', 
         fontsize=10, backgroundcolor='white', fontweight='bold')

# Q3: Low Arousal, Negative Valence
plt.text(0.05, 0.1, 'SAD / MELANCHOLIC\n(Low Energy, Low Valence)', 
         fontsize=10, backgroundcolor='white', fontweight='bold')

# Q4: Low Arousal, Positive Valence
plt.text(0.75, 0.1, 'RELAXED / CHILL\n(Low Energy, High Valence)', 
         fontsize=10, backgroundcolor='white', fontweight='bold')

# Titles and Axis Labels
plt.title('Mood Mapping: The Circumplex Model of Affect', fontsize=15)
plt.xlabel('Valence (Negativity <-> Positivity)', fontsize=12)
plt.ylabel('Energy (Low Arousal <-> High Arousal)', fontsize=12)

plt.legend(bbox_to_anchor=(1.05, 0.75), loc='upper left', title='Genre')

plt.tight_layout()
plt.show()
```

### Categorize by genre 

```python
# 'col_wrap=3' means we will have 3 charts per row
g = sns.relplot(
    data=df, 
    x='valence', 
    y='energy', 
    hue='playlist_genre', 
    col='playlist_genre', 
    col_wrap=3, 
    height=3.5, 
    aspect=1,
    alpha=0.3,      # Adjust transparency
    palette='bright'
)

# Add the Quadrant Lines to each subplot
for ax in g.axes.flatten():
    ax.axvline(0.5, color='black', linestyle='--', alpha=0.5)
    ax.axhline(0.5, color='black', linestyle='--', alpha=0.5)

g.fig.suptitle('Mood Mapping by Genre', y=1.02, fontsize=16)
plt.show()
```

### Simplify using centroids

```python
# Calculate the center (mean) for each genre
genre_means = df.groupby('playlist_genre')[['valence', 'energy']].mean().reset_index()

plt.figure(figsize=(10, 8))

# Plot backgrounds points very faintly for context
sns.scatterplot(
    data=df, x='valence', y='energy', 
    color='lightgray', alpha=0.1, s=10, label='Individual Tracks'
)

# Plot the centroids
sns.scatterplot(
    data=genre_means, 
    x='valence', y='energy', 
    hue='playlist_genre', 
    s=300, 
    marker='X', 
    edgecolor='black',
    palette='bright'
)

# Add labels to the centroids
for i in range(genre_means.shape[0]):
    plt.text(
        genre_means.valence[i]+0.02, 
        genre_means.energy[i], 
        genre_means.playlist_genre[i], 
        fontsize=12, 
        weight='bold'
    )

# Quadrant Lines
plt.axvline(0.5, color='black', linestyle='--')
plt.axhline(0.5, color='black', linestyle='--')

plt.title('Average Mood Location by Genre (Centroids)', fontsize=15)
plt.xlabel('Valence')
plt.ylabel('Energy')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

**Observation:**
The centroid analysis reveals a significant limitation in using broad genre labels to predict emotional states. As shown in the plot:

* **High Overlap:** Distinct genres such as *Pop*, *Rock*, *Latin*, and *Rap* cluster tightly in the **Euphoric Quadrant** (High Energy, High Valence), making them mathematically indistinguishable based on averages alone.
* **The Average Trap:** While the centroids suggest that "Rock is happy," the background scatter plot shows massive dispersion. All genres contains tracks across all four emotional quadrants.

**Conclusion for the User Profile Engine:**
Relying on **Genre Averages** to determine a user's psychological profile is **statistically invalid**. It forces a "one-size-fits-all" label that ignores the nuance of individual listening habits. E.g. an user listening to specific *Sad Rock* ballads would be misclassified as "Euphoric" if we only looked at the genre's average.

**Strategic Pivot:**
To build an accurate psychological profile, we must abandon genre-level aggregation and focus on **granular, track-level analysis** and **distribution spreads** (Variance/Standard Deviation) rather than simple means.

## Distribution Analysis: Measuring Emotional Versatility
Objective: To move beyond averages and evaluate the variance within each genre. This determines how reliable a genre tag is for predicting a specific audio feature.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

features = ['valence', 'energy', 'tempo']
titles = ['Emotional Range (Valence)', 'Intensity Range (Energy)', 'Rhythm Range (Tempo)']
colors = ['skyblue', 'salmon', 'lightgreen']

for i, col in enumerate(features):
    sns.boxplot(
        data=df, 
        x='playlist_genre', 
        y=col, 
        ax=axes[i], 
        palette='bright'
    )
    
    axes[i].set_title(titles[i], fontsize=14, fontweight='bold')
    axes[i].set_xlabel('')                       # Don't show labels
    axes[i].tick_params(axis='x', rotation=45)   

plt.tight_layout()
plt.show()
```

### Analysis: Decoupling Genre from Attribute
**Key Insight**: The distribution analysis reveals that **Structural features** (Tempo, Energy) are highly correlated with specific genres, while **Emotional features** (Valence) are genre-independent.

**1. Emotional Range (Valence)**

* **High Variance in Pop/Latin:** The "tall" boxes for *Latin* and *Pop* span almost the entire valence spectrum (0.2 to 0.8). This confirms that these genre tags are **emotionally agnostic**knowing a user listens to "Pop" provides zero information about whether they prefer sad or happy music.
* **The "Dark" EDM:** Surprisingly, *EDM* shows a low median valence (< 0.4), indicating that despite its high energy, the genre often relies on minor keys and tension, contrary to the "happy dance music" stereotype.

**2. Intensity Range (Energy)**

* **Genre Rigidity:** *Rock* and *EDM* exhibit "squashed" boxes at the top of the scale (Median > 0.8), showing low variance. These are **rigid genres**: if a track is classified as Rock, it is have a high change to also be high-intensity.
* **The "Chill" Factor:** *R&B* is the distinct outlier with a lower energy profile, making it the primary candidate for detecting "Relaxation" states in the user profile.

**3. Rhythm Range (Tempo)**

* **The 128 BPM Standard:** The *EDM* boxplot is virtually a flat line around 128 BPM. This is a common tempo used in parties and for physical training and is a production standard for mixability.
* **Strategic Insight:** This low variance allows for deterministic classification. A track with High Energy + High Loudness + ~128 BPM can be confidently tagged as "Club/Dance" even if the metadata is missing.

**Conclusion:**
- **Structural features** (Tempo, Energy) are highly correlated with specific genres (Rock/EDM), while **Emotional features** (Valence) are genre-independent. 
- The algorithm must therefore weight features differently depending on the context.

---

### Implications for Backend Logic & Profile Generation

Based on the distribution analysis above, we establish the following rules for the **Psychological Profile Engine**:

#### 1. The "Openness" Proxy (Variance Rule)

* **Observation:** *Pop* and *Latin* genres exhibit extreme variance in Valence and Energy (wide IQR).
* **Backend Logic:** Users with a high frequency of these genres cannot be profiled by genre tags alone. However, their consumption indicates **High Openness** (tolerance for wide emotional shifts).
* **Contrast:** Users dominated by *EDM* or *Rock* (low variance in Energy) display a preference for **Stability and Intensity**, indicating a more "Niche/Focused" listening personality.

#### 2. Feature Granularity Strategy

* **Mood Detection:** Since *Valence* is highly dispersed within all genres, the system **MUST NOT** use Genre to predict mood.
* *Action:* The API will calculate the `mood_label` (Sad vs. Happy) strictly at the **Track Level** using the specific `valence` float value of each song.
* **Intensity Detection:** *Energy* and *Tempo* are strictly clustered for *EDM* and *Rock*.
* *Action:* The system **CAN** use Genre (or a Tempo > 125 threshold) as a heuristic to tag sessions as "High Intensity" without querying every track's details.


#### 3. The "EDM Signature"

* **Rule:** If `tempo`  128 BPM AND `energy` > 0.8:
* **Classification:** The track is functionally "Electronic/Dance" regardless of its metadata. This allows us to identify "Party/Workout" states even if the genre tag is missing or incorrect.

## Pairplot analysis

```python
# We use only the important features
cols_to_plot = [
    'energy', 'valence', 'acousticness', 
    'instrumentalness', 'danceability', 'loudness'
]

# We take a sample of 1_000 songs to make it quick
sample_df = df.sample(n=1000, random_state=42)

# Generate the pairplots
plt.figure(figsize=(15, 15))
sns.pairplot(
    sample_df, 
    vars=cols_to_plot, 
    hue='playlist_genre',  
    palette='bright',
    plot_kws={'alpha': 0.5, 's': 20}, # Transparency to check superpositions
    diag_kind='kde' 
)

plt.suptitle('Pairplot Analysis: Feature Interactions & Genre Overlap', y=1.02, fontsize=16)
plt.show()
```

### Pairplot Analysis: Feature Interactions & Genre Overlap

**Objective:** To visually assess pairwise relationships between audio features and evaluate the separability of genre classes.

**Key Observations:**

1. **High Class Overlap:**
* Across most feature pairs (e.g., *Valence vs. Danceability*), the genre hues are inextricably mixed. There are no distinct "clusters" defined by genre labels.
* **Insight:** This visually confirms that **Genre is a weak proxy for musical content**. A "Pop" track can structurally resemble a "Latin" track, validating the need for track-level profiling rather than genre-level averaging.


2. **Multicollinearity Detected:**
* The scatterplot for `energy` vs. `loudness` shows a near-perfect linear diagonal.
* **Insight:** These features are redundant. For future modeling (e.g., K-Means), `loudness` adds little unique information if `energy` is present. This supports the use of dimensionality reduction (PCA) to handle this correlation.


3. **Physical Constraints (The "L-Shape"):**
* The interaction between `acousticness` and `energy` (1st column, 3rd rown) forms a strict boundary (L-shape). High acousticness dictates low energy, and vice versa. Songs with high `acousticness` and `energy` doesn't exists in the data.
* **Insight:** This mutual exclusivity could represent a primary axis of variation in the dataset ("Organic vs. Synthetic"), which could effectively separate the dataset into two physical poles.



**Conclusion:**
The lack of clear decision boundaries between genres in 2D space justifies the pivot to **Unsupervised Learning (Clustering)**. We must derive the natural groupings of the data rather than relying on pre-existing labels.
